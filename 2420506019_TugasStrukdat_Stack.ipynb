```python
class TowerOfHanoi:
    def __init__(self, jumlah_cakram, langkah_maksimal):
        self.tiang = {"A": list(reversed(range(1, jumlah_cakram + 1))), "B": [], "C": []}
        self.langkah_maksimal = langkah_maksimal
        self.langkah = 0

    def pindahkan_cakram(self, asal, tujuan):
        if self.langkah >= self.langkah_maksimal:
            print("Langkah maksimal telah tercapai!")
            return

        if not self.tiang[asal]:
            print(f"Tiang {asal} kosong, tidak bisa memindahkan cakram.")
            return

        if self.tiang[tujuan] and self.tiang[tujuan][-1] < self.tiang[asal][-1]:
            print("Pindahan tidak valid! Cakram yang lebih besar tidak boleh diletakkan di atas yang lebih kecil.")
            return

        cakram = self.tiang[asal].pop()
        self.tiang[tujuan].append(cakram)
        self.langkah += 1
        langkah_sisa = self.langkah_maksimal - self.langkah
        print(f"Langkah {self.langkah}: Memindahkan cakram {cakram} dari {asal} ke {tujuan} (Sisa langkah = {langkah_sisa})")

    def tampilkan_tiang(self):
        print("\nKeadaan Tiang:")
        for tiang, cakram in self.tiang.items():
            print(f"{tiang}: {cakram}")

    def mainkan_game(self):
        self.tampilkan_tiang()
        while True:
            asal = input("Masukkan tiang asal (A/B/C) untuk mencapai langkah maksimal atau 'X' untuk keluar: ").upper()
            if asal == 'X':
                print("Permainan dihentikan oleh pengguna.")
                break
            tujuan = input("Masukkan tiang tujuan (A/B/C) untuk mencapai langkah maksimal atau 'X' untuk keluar: ").upper()
            if tujuan == 'X':
                print("Permainan dihentikan oleh pengguna.")
                break

            if asal in self.tiang and tujuan in self.tiang:
                self.pindahkan_cakram(asal, tujuan)
                self.tampilkan_tiang()
            else:
                print("Input tidak valid. Coba lagi.")

            if self.langkah >= self.langkah_maksimal:
                print("Game over! Langkah maksimal tercapai.")
                break

        if self.tiang["C"] == list(reversed(range(1, len(self.tiang["C"]) + 1))):
            print("Selamat! Anda berhasil menyelesaikan Tower of Hanoi.")


# Contoh Penggunaan Game
# Dua digit terakhir digunakan sebagai batas langkah
jumlah_cakram = 3
langkah_maksimal = 19

game = TowerOfHanoi(jumlah_cakram, langkah_maksimal)
game.mainkan_game()
```
